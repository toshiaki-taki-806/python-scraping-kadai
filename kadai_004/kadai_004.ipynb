```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import datetime
import time

def extract_stock_data(source):
    stock_data_list = []
    soup = BeautifulSoup(source, 'html.parser')
    for i in range(5):
        stock_data = soup.select_one(f'#highcharts-4 > div.highcharts-tooltip > span > table > tbody > tr:nth-child(1) > td:nth-child({i + 1})')
        stock_value = stock_data.text
        #print(stock_value)
        if i == 0:
            stock_date = datetime.datetime.strptime(stock_value, '%Y/%m/%d').strftime('%Y/%m/%d')
            stock_data_list.append(stock_date)
        else:
            stock_data_list.append(stock_value[4:])
    return stock_data_list

def get_stcok_values(chrome_driver, url):
    chrome_driver.get(url)
    wait = WebDriverWait(chrome_driver, 20)
    wait.until(EC.presence_of_all_elements_located)

    # グラフの要素を取得
    graph = chrome_driver.find_element(By.XPATH, '//*[@id="JSID_chart"]')

    # マウスの動きを記述
    actions = ActionChains(chrome_driver)

    # グラフの中心にマウスを移動
    actions.move_to_element(graph)

    # グラフの横幅（width）の半分の値だけ右にマウスをずらし、グラフの右端にマウスを移動
    g_width = graph.rect['width']
    actions.move_by_offset((g_width // 2),0)

    # 記述した動作の実行
    actions.perform()

    for i in range(g_width):
        print(i)
        if i != 0:
            # マウスを左に移動させる処理
            actions = ActionChains(chrome_driver)
            actions.move_by_offset(-1,0)
            actions.perform()

        #htmlを取得して株価を取得
        html_source = chrome_driver.page_source
        stock_all_list.append(extract_stock_data(html_source))

if __name__ == '__main__':

    #初期化
    html_source = ""
    stock_all_list = []
    #設定
    url = 'https://www.nikkei.com/markets/worldidx/chart/nk225/?type=6month'
    # ヘッドレスモードで起動するためのオプションを設定
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    # Chromeを立ち上げる
    chrome_driver = webdriver.Chrome(options=chrome_options)

    start_time = time.time()
    get_stcok_values(chrome_driver, url)
    end_time = time.time()
    time_diff = end_time - start_time
    print(f'実行時間: {time_diff}')
    print(stock_all_list)
    chrome_driver.quit()
```

```python
#データ成型
import pprint

print(f'実行時間：{time_diff}秒')
def get_unique_list(seq):
    seen = []
    return [x for x in seq if x not in seen and not seen.append(x)]
stock_unique_list = get_unique_list(stock_all_list)
print(f'データ取得数：{len(stock_unique_list)}')
print('日付,始値,高値,安値,終値')
pprint.pprint(stock_unique_list)
```
